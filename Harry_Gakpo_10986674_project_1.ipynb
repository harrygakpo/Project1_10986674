### To show data correlation among columns, I used the corr method in the following code.

```python
matrix = df.corr().round(2)
```

```python
print(matrix)
```

### This is a heatmap showing the correlation among columns.

```python
sns.heatmap(matrix, annot=True)
plt.show()
```

## HIDDEN INFERENCES FROM THE HEATMAP DIAGRAM

###  1. From the heatmap diagram above, the correlations between the various columns can be easily inferred. A deep blue infers a strong positive correlation. As the deepness of the blue color decreases, it infers a weak correlaton as it approaches negative.
### 2. From the diagram, it is clearly seen that every column correlates at the maximum positive correlation of 1 against its own colum  and that is reasonable as the values are the same.
###  3. Also it can be clearly observed that the correlation between fixed acidity and pH approaches a minimum correlation as their correlation depicts a very very light color
### 4. If we take a critical look at the heat map, it can be observed that  free sulfur dioxide and total sulfur dioxide have a pretty strong positive correlation as their correlation blue color is quite deep depicting a strong correlation
